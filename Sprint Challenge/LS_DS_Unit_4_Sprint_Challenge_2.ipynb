## 1. Define the following terms:

- **Neuron:**
    A Neuron is a single node within a Neural Network.
- **Input Layer:**
    The input layer is the highest level layer, where we input our data into the Neural Network
- **Hidden Layer:**
    Hidden layers are generally called that because we really don't have much access to them. 
    They're the layers where a lot of the neural network magic happens, crunching ming numbing
    amounts of numbers in a fraction of the time our minds could work through them.
- **Output Layer:**
    The output layer is where we see what our Neural Network comes up with. In some cases
    this means that we see a prediction of something. In others, a new thing, based on a lot
    of old things.
- **Activation:**
    Activation in a neural network is like a neuron firing! It's a big part of the math that 
    allows a feed forward Neural Network to make sense of the relationships in data.
- **Backpropagation:**
    Backpropagation is a method of sending all the information learned, backwards in the network.
    The process updates our original weights, allowing the model to fit even more effectively to the
    data.

```python
#Defining a sigmoid function for my perceptron
def sigmoid(x):
    return 1/ (1 + np.exp(-x))
```

```python
#Establish a set of random weights
weights = np.random.random((2,10000))
weights
```

```python
#Impliment basic perceptron calculate weighted sum
weighted_sum = np.dot(X, weights)
#calculate acticvated output of perceptron(prediction)
activated_output = sigmoid(weighted_sum)
print(weighted_sum)
print(activated_output)
```

```python
from sklearn.metrics import mean_squared_error
#Calculate error
print(mean_squared_error(y, activated_output))
```

```python
#Define sigmoid derivative to adjust weights with each epoch
def sigmoid_derivative(x):
    sx = sigmoid(x)
    return sx * (1 - sx)
```

```python
for iteration in range(100):
    #Calculate input times weights
    weighted_sum = np.dot(X, weights)
    #Run through activation function
    activated_output = sigmoid(weighted_sum)
    #Calculate the error
    error = y - activated_output
    #Calculate adjustments
    adjustments = error * sigmoid_derivative(activated_output)
    #Adjust weights
    weights += np.dot(X.T, adjustments)
    print("Epoch Complete")
    
print(f"Weights after training: {weights}")
print(f"Output after training: {activated_output}")
```

### Multilayer Perceptron <a id="Q3"></a>

Using the sample candy dataset, implement a Neural Network Multilayer Perceptron class that uses backpropagation to update the network's weights. Your Multilayer Perceptron should be implemented in Numpy. 
Your network must have one hidden layer.

Once you've trained your model, report your accuracy. Explain why your MLP's performance is considerably better than your simple perceptron's on the candy dataset. 

```python
class NeuralNetwork:
    def __init__(self):

        #Establish neural network architecture
        #Specify number of nodes in each layer
        
        self.inputs = 2
        self.hiddenNodes = 3
        self.outputNodes = 10000

        #Randomly initialize weights input layer -> hidden layer
        self.weights1 = np.random.rand(self.inputs, self.hiddenNodes)
       
        #Random intitialize weights hidden layer -> output layer
        self.weights2 = np.random.rand(self.hiddenNodes, self.outputNodes)
        
    def sigmoid(self, x):
        #Adding sigmoid function to actual class
        return 1/ (1 + np.exp(-x))
    
    def sigmoidPrime(self,z):
        #Gradient of sigmoid
        return np.exp(-z)/((1+np.exp(-z))**2)
        
    def feed_forward(self, X):
        """
        impliment above logic to create the feed forward portion
        """
       
        #Combining inputs and weights in a weighted sum
        self.hidden_sum = np.dot(X, self.weights1)
        
        #Apply activation function to the hidden layer weight sums. 
        #Activated values
        self.activated_hidden = self.sigmoid(self.hidden_sum)
        
        #Weighted sum hidden layer and the output
        self.output_sum = np.dot(self.activated_hidden, self.weights2)
        
        #Apply activation function to the output weighted sums
        self.activated_output = self.sigmoid(self.output_sum)
        
        return self.activated_output
    
    def backward(self, X,y,o):
        """
        Send the new information back through the network
        """
        
        #Calculate the error
        self.o_error = y - o
        
        #Apply sigmoid derivative to the error. 
        #Which direction do we want go?
        self.o_delta = self.o_error * self.sigmoidPrime(o)
        
        #How much our hidden layer weights contributed to output error.
        self.z2_error = self.o_delta.dot(self.weights2.T)
        
        #Apply sigmoid derivative to the error. 
        self.z2_delta = self.z2_error * self.sigmoidPrime(self.activated_hidden)
        
        #Applying adjustments to the weights.
        self.weights1 += X.T.dot(self.z2_delta)
        
        #Applying adjustments to the weights.
        self.weights2 += self.activated_hidden.T.dot(self.o_delta)
        
    def train(self, X, y):
        #Simple training function from class.
        o = self.feed_forward(X)
        self.backward(X,y,o)
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler

df = pd.read_csv('https://raw.githubusercontent.com/ryanleeallred/datasets/master/heart.csv')
df = df.sample(frac=1)
print(df.shape)
df.head()
```

```python
#Import Keras parts
from sklearn.model_selection import train_test_split
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.wrappers.scikit_learn import KerasClassifier
```

```python
#Establish build function for KerasClassifier
def create_model():
    model = Sequential()
    model.add(Dense(10, input_dim=13, activation="relu"))
    model.add(Dense(16, activation="sigmoid"))
    model.add(Dense(1, activation="sigmoid"))
    model.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
    return model
```

```python
kc = KerasClassifier(build_fn=create_model, verbose=1)
```

```python
X = df.drop("target", axis=1)
y = df["target"]
```

```python
#Establish X/y test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33)
#Establish X/y train and validation sets
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.33)
```

```python
#Initial baseline fit
kc.fit(X_train, y_train)
```

```python
#adding epochs
kc.fit(X_train, y_train, epochs=50)
```

```python
from sklearn.model_selection import GridSearchCV

param_grid = {'batch_size':[1,5,10],
              'epochs':[15,30,45]}

grid = GridSearchCV(estimator=kc, param_grid=param_grid)
```

```python
results = grid.fit(X_train, y_train)
```
